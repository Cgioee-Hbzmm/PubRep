```python
def combinations(n, k):
    if k == 0:
        return [[]]
    if k > n:
        return []

    if n == k:
        return [list(range( n ))]

    # Compute combinations by including or excluding the last element
    include_last = combinations(n - 1, k - 1)
    exclude_last = combinations(n - 1, k)

    # Add the last element to the included combinations
    include_last = [comb + [n-1] for comb in include_last]

    # Combine the two sets of combinations
    return include_last + exclude_last

# Example usage:
n = 6
k = 2
result = combinations(n, k)
result
```

```python
from itertools import product,combinations

# Define your 8-tuple x and list of 8 primes p
p = [2, 3, 5, 7, 11, 13, 17, 19]  # Replace these with your actual prime numbers
def CRTGen8(x):
  # Get user input for the positions that are common (1-based indexing)
  valid_combinations = []
  for i in combinations(list(range(8)),4):
    com_pos=i

  # Calculate the positions that vary
    vary_positions = [i for i in range(8) if i not in com_pos]

  # Initialize a list to store the valid combinations of c

  # Iterate through the possible values for the varying positions
    for values in product(*[range(p[i]) for i in vary_positions]):
      # Construct the potential c tuple with common values from x and varying values
      c = [x[i] if i in com_pos else values[vary_positions.index(i)] for i in range(8)]

      # Append c to the list of valid combinations
      valid_combinations.append(tuple(c))

  # Print the valid combinations
  return (set(valid_combinations))

x = [0]*8
Sx=CRTGen8(x)
len(Sx)
```

```python
p = [2, 3, 5, 7, 11]
from itertools import product,combinations
def CRTGen5(x):
  valid_combinations = []
  for i in combinations(list(range(5)),3):

    var_pos=i
    com_pos = [a for a in range(5) if a not in var_pos]

    for val in product(*[range(p[i]) for i in var_pos]):

      c = [x[i] if i in com_pos else val[var_pos.index(i)] for i in range(5)]
      valid_combinations.append(tuple(c))

  return set(valid_combinations)

x = [0]*5
Sx=CRTGen5(x)

len(Sx)
```
